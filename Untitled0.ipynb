```python
headers = ['name', 'title', 'dept', 'salary']
salaries = pd.read_csv('city-of-chicago-salaries.csv', 
                     header=False, 
                     names=headers, 
                     converters={'salary': lambda val: float(val.replace('$', ''))})
salaries.head()
```

```python
by_dept = salaries.groupby('dept')
```

```python
print salaries.ix[[1,2,10]]
salaries.ix[by_dept.groups['AVIATION']].head()
```

```python
by_dept.size() # Total Records
by_dept.count() # Non-Null records in each field
```

```python
#print by_dept.sum(), '\n'
#print by_dept.mean(), '\n'
#print by_dept.median(), '\n'
print by_dept.min().salary.order(), '\n'
```

```python
def ranker(df):
    df['dept_rank'] = np.arange(len(df))
    return df
salaries.sort('salary', ascending=False, inplace=True)
group = salaries.groupby('dept').apply(ranker)
print group[group.dept_rank == 0].head(5)
```
